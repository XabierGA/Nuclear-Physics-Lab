```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from scipy.optimize import curve_fit
sns.set()
```

```python
channel = [1,2,3,4,5]
freq= [2,1,1,0,3]

plt.plot(channel, freq , 'bo')
plt.xlabel('Channel Number')
plt.ylabel('Frequency')

plt.savefig('channeldiag.pdf')


```

```python
distance = [0,5, 10 , 15 , 20 , 25 , 28]
x = [(dis+11) for dis in distance]
energies = [125.95,110.74,91.19,77.37,52.56,27.17,11.34]
def linear_c(x,a,b):
    return a*x+b
coeff , var = curve_fit(linear_c , x , energies)
a = coeff[0]
b = coeff[1]
step = np.linspace(0 , 40 , 1000)
pl = []
for i in range(len(step)):
    pl.append(step[i]*a + b)
    
plt.plot(step, pl , 'b-' , label = 'Linear Fit')
plt.plot(x, energies , 'mo' , label = 'Experimental Points')
plt.ylabel('Energies (keV)')
plt.xlabel('Distance (mm)')
plt.legend()

plt.savefig('enerdisfit.pdf')
```

```python
std = np.sqrt(np.diag(var))
print(coeff)
print(std)
```

```python
count = [85.43,37.64,19.37,13.21,8.76,6.90,1.09]
plt.plot(x, count , 'mo' , label = 'Experimental Points')
plt.ylabel('Count Rate')
plt.xlabel('Distance (mm)')
def nonlinear_c(x,c,d):
    return c/(x**2) + d
params = np.array([1,1])
coeff1 , var1 = curve_fit(nonlinear_c , x , count , params)
a1 = coeff1[0]
b1 = coeff1[1]
step1 = np.linspace(11 , 39 , 1000)
pl1 = []
for i in range(len(step1)):
    pl1.append(a1/(step1[i]**2) + b1)
plt.plot(step1, pl1 , 'b-', label = 'Non linear fit')
plt.legend()
plt.savefig('countdis.pdf')
```

```python
plt.plot(step1, pl1 , 'bo')
print(a1)
print(b1)
```
